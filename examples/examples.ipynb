```python
it = 0
for label in labels:
    fields = bulk_fields[label] * -14.4
    counts, bins = np.histogram(fields, range = (-5, 1), bins=101)
    binw = bins[1] - bins[0]
    bin_centers = bins[:-1] + binw/2

    plt.plot(bin_centers, counts)
    it += 0


plt.xlabel("E (V/A)", fontsize=15)
plt.ylabel("Counts", fontsize=15)
```
