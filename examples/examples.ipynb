```python
import warnings
from sklearn.exceptions import InconsistentVersionWarning

# The bundled models in data/ were trained with an older scikit-learn.
# Silence the version-mismatch warnings before importing model, which
# loads those models at import time.
warnings.filterwarnings("ignore", category=InconsistentVersionWarning)

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import sys
sys.path.insert(0, '..')

from model import compute_frequencies
```

```python
# Normally you would feed in a big list of electric fields computed from MD simulation
# Here we just use a small range, showing its affect on the symmetric stretch

field_strengths = np.linspace(0, -2, 11)
example_fields = np.zeros([11, 2])
example_fields[:,0] = field_strengths
```

```python
compute_frequencies(example_fields, "../amoeba_params.json")
```

```python
results = np.load("results.npz")
```

```python
# For 'freqs', Each column is four lowest frequencies. For gas-phase water, that is
# 1) Bending Fundamental
# 2) Bending Overtone
# 3) Symmetric Stretch
# 4) Asymmetric Stretch
freqs = results['freqs']

# activities contains the Raman an IR activites, in this order
# 0-3)  Raman Isotropic
# 4-7)  Raman Anisotropic
# 8-11)  IR One-body; not realistic, only useful for debugging
# 12-15)  IR Two-body; actual IR
activities = results['activities']


plt.scatter(field_strengths, freqs[:,2])
plt.xlabel("Electric fields, (V/A)", fontsize=15)
plt.ylabel("Frequencies (cm $^{-1}$)", fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()

plt.scatter(field_strengths, activities[:,14])
plt.xlabel("Electric fields, (V/A)", fontsize=15)
plt.ylabel("Sym. Stretch IR activivty (a.u.)", fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```
